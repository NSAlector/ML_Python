```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("KNNAlgorithmDataset.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop("Unnamed: 32", axis=1, inplace=True)
```

```python
plt.hist(df['diagnosis'], bins = 2)
```

```python
df=pd.get_dummies(df,{'diagnosis'}) #кодировка категориальных признаков
```

```python
df
```

```python
df['diagnosis_B'] =  pd.to_numeric(df['diagnosis_B'])
```

```python
df['diagnosis_M'] =  pd.to_numeric(df['diagnosis_M'])
```

```python
df
```

```python
df['diagnosis_M'] = df['diagnosis_M'].astype(int)
```

```python
df['diagnosis_B'] = df['diagnosis_B'].astype(int)
```

```python
df
```

```python
df.info()
```

```python
X=df.drop(['id','diagnosis_B','diagnosis_M'],axis=1)
```

```python
y=df['diagnosis_M']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8,test_size=0.2, random_state=11)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gauss_nb = GaussianNB()
gauss_nb.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
pred_train=gauss_nb.predict(X_train)
print(classification_report(y_train,pred_train))
```

```python
pred_test=gauss_nb.predict(X_test)
print(classification_report(y_test,pred_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=1)
KNN.fit(X_train, y_train)
```

```python
print(classification_report(y_train,KNN.predict(X_train)))
```

```python
print(classification_report(y_test,KNN.predict(X_test)))
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
KNN.fit(X_train, y_train)
```
